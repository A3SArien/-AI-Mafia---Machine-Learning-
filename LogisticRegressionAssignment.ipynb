# Assignment 3 - Chemicals Segregation
Can you Separate the chemicals?

Binary Classification Challenge

A chemist has two chemical flasks labelled 1 and 0 which contains two different chemicals.

He extracted 3 features from these chemicals in order to distinguish between them. You are provided with the results derived by the chemist and your task is to create a model that will label chemical 0 or 1 given its three features.

## Data Description
You are provided with two files test and train.
-  Train: This files consists of two csv files Logistic Xtrain and LogisticYtrain. Xtrain
consists of the features whereas Ytrain consists of the labels associated with the features.
- Test: This file consists of two files Logistic Xtest consisting of the features of test
data

Importing lib and datasets

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings(action = 'ignore')
```

```python
x_train = pd.read_csv('/content/drive/MyDrive/ML: Codes and Assignments/Week 2 - Assignments (Linear and Logistic Regression)/Logistic Regression/Training Data/Logistic_X_Train.csv')
x_test = pd.read_csv('/content/drive/MyDrive/ML: Codes and Assignments/Week 2 - Assignments (Linear and Logistic Regression)/Logistic Regression/Testing Data/Logistic_X_Test.csv')
y_train = pd.read_csv('/content/drive/MyDrive/ML: Codes and Assignments/Week 2 - Assignments (Linear and Logistic Regression)/Logistic Regression/Training Data/Logistic_Y_Train.csv')
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
```

```python
x_train = ss.fit_transform(x_train)
x_test = ss.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
mod = lr.fit(x_train,y_train)
```

```python
y_pred = mod.predict(x_test)
```

```python
mod.score(x_train,y_train)
```

```python
mod.score(x_test,y_pred)
```

```python
mod.decision_function(x_test)
```
