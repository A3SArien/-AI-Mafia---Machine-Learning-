```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# Training

model.fit(x,y)
```

```python
#Prediction

pred = model.predict(x)
```

```python
model.score(x,y)
```

```python
## Visualise

plt.scatter(x,y,label='data')
plt.plot(x,pred,color='red',label='prediction')
plt.legend()
plt.show()
```
